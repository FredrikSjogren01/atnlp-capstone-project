```python
#TRY
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from rank_bm25 import BM25Okapi

# Directory containing the text files
directory = '/content/triviaqa-rc/evidence/wikipedia'

# Function to remove stop words from a question
def stop_word_removal(question):
    tokens = word_tokenize(question)
    return [token for token in tokens if token.lower() not in stopwords.words('english')]

# Function to extract nouns and verbs from a question
def find_nouns_verbs(question):
    tokenized = nltk.word_tokenize(question)
    pos_tags = nltk.pos_tag(tokenized)
    nouns_and_verbs = [word for (word, pos) in pos_tags if pos[:2] == 'NN' or pos[:2] == 'VB']
    names = [word for (word, pos) in pos_tags if pos in ['NNP', 'NNPS']]
    return nouns_and_verbs, names

# Function to retrieve relevant chunks using cached BM25 model
def retrieve_relevant_chunks_cached(bm25, query, chunks, min_score):
    # Tokenize the query once
    tokenized_query = word_tokenize(query.lower())
    # Retrieve relevance scores from the cached BM25 model
    scores = bm25.get_scores(tokenized_query)
    return [chunks[i] for i in range(len(scores)) if scores[i] > min_score]

# Function to tokenize all chunks and cache the BM25 model
def prepare_bm25_model(chunks):
    # Tokenize all chunks once
    tokenized_chunks = [word_tokenize(chunk.lower()) for chunk in chunks]
    # Initialize BM25 model with tokenized chunks
    bm25 = BM25Okapi(tokenized_chunks)
    return bm25

# Main function to extract relevant chunks
def get_relevant_chunks(question, min_score=2, max_files=30):
    relevant_chunks = []
    nouns_in_question, names = find_nouns_verbs(question)
    file_count = 0

    # List to store all chunks for BM25
    all_chunks = []

    # Loop through text files in the directory
    for filename in os.listdir(directory):
        if filename.endswith('.txt'):
            filepath = os.path.join(directory, filename)
            with open(filepath, 'r') as file:
                text = file.read()

            # Split the document into chunks
            text_chunks = text.split('\n\n')
            all_chunks.extend(text_chunks)  # Add chunks to the list

            # Increment file counter and stop if max_files limit is reached
            file_count += 1
            if file_count >= max_files:
                break

    # Prepare BM25 model once for all chunks
    bm25 = prepare_bm25_model(all_chunks)

    # Now, process each query and retrieve relevant chunks using the cached BM25 model
    filtered_chunks = retrieve_relevant_chunks_cached(bm25, question, all_chunks, min_score)

    # Keep only chunks that contain at least one noun from the question
    for chunk in filtered_chunks:
        tokenize_chunk = set(word_tokenize(chunk.lower()))
        if names:
            if set(map(str.lower, names)).intersection(tokenize_chunk):
                relevant_chunks.append(chunk)
        else:
            if set(nouns_in_question).intersection(tokenize_chunk):
                relevant_chunks.append(chunk)

    return relevant_chunks
```

```python
def can_answer_question(question, context):
    # Format the input for the T5 model
    input_text = f"question: {question} context: {context}"
    inputs = tokenizer(input_text, return_tensors="pt", padding=True, truncation=True)

    # Generate the response
    outputs = model.generate(inputs['input_ids'], max_length=5, num_beams=4, early_stopping=True)

    # Decode and extract the response
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Check if the response is 'yes' or 'no'
    return response.lower() in ['yes', 'no']

# Example usage
question = "Can I play soccer in the rain?"
context = "Soccer can be played in the rain, although it may be harder to control the ball."

# Check if the question can be answered
can_be_answered = can_answer_question(question, context)
print("Can the question be answered?", can_be_answered)
```
